```python
def download_galah_fits(catalog_summary, output_dir="/datax/scratch/emmay/galah_spectra"):
    """
    Download GALAH FITS files for up to 100 objects with is_in_GALAH == 1
    for B, V, R, I filters, and save as objid_filter.fits if not already present.
    
    Parameters:
        catalog_summary (pd.DataFrame): DataFrame with at least 'GALAH_name' and 'is_in_GALAH' columns
        output_dir (str): Directory to save downloaded files
    """

    os.makedirs(output_dir, exist_ok=True)

    galah_objects = catalog_summary[catalog_summary['is_in_GALAH'] == 1]
    filters = ['B', 'V', 'R', 'I']
    base_url = "https://datacentral.org.au/vo/slink/links"

    for _, row in galah_objects.iterrows():
        obj_id = str(int(row['GALAH_name']))

        for filt in filters:
            filename = f"{obj_id}_{filt}.fits"
            filepath = os.path.join(output_dir, filename)

            # Skip download if file already exists
            if os.path.exists(filepath):
                print(f"Already exists: {filename}, skipping.")
                continue

            params = {
                "ID": obj_id,
                "DR": "galah_dr4",
                "FILT": filt,
                "RESPONSEFORMAT": "fits"
            }
            response = requests.get(base_url, params=params)

            if response.status_code == 200:
                with open(filepath, 'wb') as f:
                    f.write(response.content)
                print(f"Downloaded: {filename}")
            else:
                print(f"Failed to download {filename} — HTTP {response.status_code}")
```

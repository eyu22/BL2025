```python
# Open SIMBAD cross matches with GCNS
with fits.open('/datax/scratch/emmay/GCNS_SIM/simbad_gcns_crossmatches') as hdul:
    data = hdul[1].data

# Convert to Pandas for easy cleaning
df = pd.DataFrame(data.tolist(), columns=data.columns.names)
```

```python
# Make sure the column name is exactly 'main_id'
df['main_id'] = df['main_id'].astype(str).str.strip()

# Save to text file
with open("gcns_names.txt", "w") as f:
    for name in df['main_id'].dropna():
        f.write(f"{name}\n")
```

```python
# Read all lines from the file
with open("gcns_coords.txt", "r") as f:
    lines = f.readlines()

# Total lines and size of each chunk
total = len(lines)
chunk_size = total // 10 + (total % 10 > 0)  # round up to cover all lines

# Split and save
for i in range(10):
    start = i * chunk_size
    end = start + chunk_size
    chunk = lines[start:end]

    with open(f"gcns_coords_part{i+1}.txt", "w") as f_out:
        f_out.writelines(chunk)
```
